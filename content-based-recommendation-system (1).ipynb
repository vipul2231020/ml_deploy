# Content Based recommendation system

## Required Feilds

```python
data = pd.read_csv('data.csv')
```

```python
data = data[['Eid', 'Name', 'Experience', 'Total_projects', 'Area_of_Interest_1']]
```

```python
data
```

### Joim four coloumn for vectorization

```python

data['Experience'] = pd.to_numeric(data['Experience'], errors='coerce')
data['Total_projects'] = pd.to_numeric(data['Total_projects'], errors='coerce')


data['newcolm'] = (data['Experience'] + data['Total_projects']).astype(str) + " " + data['Area_of_Interest_1']+" "+data['Name']
```

```python
data
```

### import vectorization and cosine similarity 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfif =TfidfVectorizer()
matrix = tfif.fit_transform(data['newcolm'])
similarity = cosine_similarity(matrix)
```

```python
similarity
```

```python
similarity[0]
```

When user give the domain as per his requirements this function will recommend him persons suitable for the post on the basis of the experience and number of projects.  

```python


def recommendation(domain, data, matrix, similarity):
    if domain not in data['Area_of_Interest_1'].values:
        print(f"Domain '{domain}' not found in the dataset.")
        return
    domain_data = data[data['Area_of_Interest_1'] == domain]

    domain_indices = domain_data.index
    idx = domain_indices[0]
    similarity_scores = similarity[idx]
    similar_indices = np.argsort(similarity_scores)[::-1]

    print(f"Top 20 recommended employees for '{domain}':")
    for i in similar_indices[:20]:
        recommended_employee_name = data.loc[i, 'Name']
        recommended_employee_experience = data.loc[i, 'Experience']
        recommended_employee_projects = data.loc[i, 'Total_projects']
        print(f"Name: {recommended_employee_name}, Experience: {recommended_employee_experience}, Total Projects: {recommended_employee_projects}")
```

## Recommendation for the given input 

Example: user fill Domain ML 
and it will receive 20 suggestion on the basis of domain,experience and total number of projects

```python
recommendation("Backend", data, matrix, similarity)
```
